### LSTM are sensitive to scale of data.so we apply MinMax scaler

```python
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df1
```

```python
np.set_printoptions(suppress=True)
df1
```

```python
training_size = int(len(df1) * 0.70)
train_data = df1[:training_size]
test_data = df1[training_size:]
```

```python
training_size
```

```python
len(train_data)
```

```python
len(test_data),test_data
```

```python
import numpy
def create_dataset(dataset, time_step):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step):
        dataX.append(dataset[i:(i + time_step), 0])
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
#reshape into X=t,t+1,t+2,t+3  and Y=t+4
time_step=100
X_train, y_train=create_dataset(train_data, time_step)
X_test, ytest=create_dataset(test_data, time_step)
```

```python
print(X_train)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("ytest shape:", ytest.shape)
```

```python
# reshape input to be [samples,time steps,features] which is required for LSTM
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python
### Create the  Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_test, ytest), epochs=100, batch_size=64, verbose=1)
```

```python
import tensorflow as tf
tf.version
```

```python
## Do prediction and check performance matrix 
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
#transformback to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
## Root mean square error performance matrix
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
## Test Data RMSE
math.sqrt(mean_squared_error(ytest,test_predict))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create empty arrays same shape as df1
trainPredictPlot = np.empty_like(df1)
trainPredictPlot[:, :] = np.nan

testPredictPlot = np.empty_like(df1)
testPredictPlot[:, :] = np.nan

# Place train predictions after `time_step` index
trainPredictPlot[time_step:training_size, :] = train_predict

# Place test predictions after training_size + time_step
testPredictPlot[training_size + time_step:len(df1), :] = test_predict

# ⬇️ Plot everything
plt.figure(figsize=(10, 4))  # Compressed graph

plt.plot(scaler.inverse_transform(df1), label='Actual', linewidth=1)
plt.plot(trainPredictPlot, label='Train Prediction', linewidth=1)
plt.plot(testPredictPlot, label='Test Prediction', linewidth=1)

plt.legend(loc='upper left')
plt.grid(True)
plt.title("LSTM Prediction vs Actual Stock Price")
plt.xlabel("Time Step")
plt.ylabel("Scaled Price")
plt.tight_layout()
plt.show()
```
